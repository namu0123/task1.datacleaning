# Titanic Dataset - Exploratory Data Analysis (EDA)

## Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```
df = sns.load_dataset('titanic')
df.head()
```

## Dataset Info

```
df.info()
```

## Statistical Summary

```
df.describe()
```

## Survival by Gender

```
sns.countplot(x='survived', hue='sex', data=df)
plt.show()
```
